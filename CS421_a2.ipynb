## Q2 Prepare for training RNNs (10)
In this part, you are required to define the tokenizers for english and french, tokenize the data, and define the dataloaders.

Choose and initialize the tokenizer

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('google-bert/bert-base-multilingual-cased') # MULTILINGUAL MODEL such as https://huggingface.co/google-bert/bert-base-multilingual-cased
tokenizer
````

You will need to create a pytorch dataset to process the tokens in the required format. Complete the implementation of the dataset.

````python
from torch.utils.data import Dataset

class TranslationDataset(Dataset):
    def __init__(self, dataset, input_size, output_size):
        source_texts = [text["translation"]['fr'] for text in dataset]
        target_texts = [text["translation"]['en'] for text in dataset]
        self.source_sentences = tokenizer(source_texts, padding='max_length', truncation=True, return_tensors="pt")["input_ids"].cuda()
        self.target_sentences = tokenizer(target_texts, padding='max_length', truncation=True, return_tensors="pt")["input_ids"].cuda()
        self.input_size = input_size  # Not really used
        self.output_size = output_size

    def __len__(self):
        return len(self.source_sentences)

    def __getitem__(self, idx):
        return self.source_sentences[idx], self.target_sentences[idx]
````

Initialize the datasets

````python
train_dataset_rnn = TranslationDataset(train_dataset, 10200, 10200)
eval_dataset_rnn = TranslationDataset(eval_dataset, 1800, 1800)
test_dataset_rnn = TranslationDataset(test_dataset, 3000, 3000)
````

Get the vocab size from the tokenizer

````python
vocab_size = tokenizer.vocab_size # This size is used in the model instantiation
````

Initialize and define the dataloaders

````python
#Instantiate the DataLoaders
from torch.utils.data import DataLoader
BATCH_SIZE = 8 # usually 32
train_dataloader = DataLoader(train_dataset_rnn, batch_size=BATCH_SIZE, shuffle=True) # shuffled for testing
eval_dataloader = DataLoader(eval_dataset_rnn, batch_size=BATCH_SIZE)
test_dataloader = DataLoader(test_dataset_rnn, batch_size=BATCH_SIZE)
````

## Q3: Implementing RNNs (10)
Define the RNN model as an encoder-decoder RNN for the task of translation in the cell below. You may refer: https://pytorch.org/tutorials/intermediate/seq2seq_translation_tutorial.html

````python
import torch
import torch.nn as nn
````

````python
class Seq2SeqRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Seq2SeqRNN, self).__init__()
        self.hidden_size = hidden_size
        self.embedding = nn.Embedding(input_size, hidden_size) # Embedding to convert into vector
        self.encoder_gru = nn.GRU(hidden_size, hidden_size, batch_first=True) # Encoding
        self.decoder_gru = nn.GRU(hidden_size, hidden_size, batch_first=True) # Decoding which takes encoded output as input
        self.out = nn.Linear(hidden_size, output_size) # Linear layer

    def forward(self, x):
        # encoding
        embedded = self.embedding(x)
        encoder_outputs, encoder_hidden = self.encoder_gru(embedded)
        # decoding
        decoder_output, _ = self.decoder_gru(encoder_outputs, encoder_hidden)
        output = self.out(decoder_output)

        return output
````

````python
model = Seq2SeqRNN(input_size = vocab_size, hidden_size= 256, output_size = vocab_size)
model
````

## Q4: Training RNNs (15)
In this question, you will define the hyperparameters, loss and optimizer for training. You will then implement a custom training loop.

````python
if torch.cuda.is_available():
    model.cuda()
````

````python
if torch.cuda.is_available():
  print('true')
````

define the optimizer and the loss function

````python
from torch.optim import Adam
from torch.nn import CrossEntropyLoss
NUM_EPOCHS = 3
num_train_epochs = NUM_EPOCHS
num_training_steps = num_train_epochs * len(train_dataloader)
criterion = CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=0.001)
````

````python
model.to('cuda')
````

````python
import torch, gc
gc.collect()
torch.cuda.empty_cache()
````

Write the training loop

````python
from tqdm import tqdm
progress_bar = tqdm(total=num_training_steps, desc="Training Progress")

for epoch in range(num_train_epochs):
    # Training Phase
    model.train()
    total_loss = 0
    for batch_src, batch_tgt in train_dataloader:
        # sending data to cuda
        if torch.cuda.is_available():
          batch_src.to('cuda')
          batch_tgt.to('cuda')

        ## Complete the training loop
        optimizer.zero_grad()
        # This calls the forward() function within the model
        model_output = model(batch_src)

        # Here, we compute the loss function which takes in model's output and the actual target sentences of current batch
        loss = criterion(model_output.view(-1,model_output.size(-1)), batch_tgt.view(-1))

        # Backward Pass
        loss.backward()
        # Controlling the learning rate
        optimizer.step()


        progress_bar.update(1)

    # Evaluation Phase
    model.eval()

    total_batches = 0
    with torch.no_grad():
      for batch_src, batch_tgt in eval_dataloader:
        # sending data to cuda
        if torch.cuda.is_available():
          batch_src.to('cuda')
          batch_tgt.to('cuda')
        eval_outputs = model(batch_src)
        loss = criterion(eval_outputs.view(-1,eval_outputs.size(-1)), batch_tgt.view(-1))

        # Converts loss to a Python number - easier to work with - ideally for eval
        total_loss += loss.item()

        total_batches += 1

      ### Complete the evaluation phase

    avg_loss = total_loss/total_batches
    print(f"Epoch {epoch}: Average Eval Loss: {avg_loss:.4f}")
````

## Q5: Evaluating RNNs for Machine Translation (5)

Implement the calculation of BLEU-1,2,3,4 scores using the ```sacrebleu``` library for the test dataset.

````python
import evaluate
model.eval()
bleu1, bleu2, bleu3, bleu4 = None, None, None, None

progress_bar = tqdm(total=len(test_dataloader), desc="Testing Progress")
preds = []
targets = []

for batch_src, batch_tgt in tqdm(test_dataloader):
    if torch.cuda.is_available():
      batch_src.to('cuda')
      batch_tgt.to('cuda')

    output = model(batch_src)
    out_seq = torch.argmax(output, dim=-1).squeeze().tolist()
    # print(out_seq)
    # batch = {k: v.to(model.device) for k, v in batch.items()}
    outs = tokenizer.batch_decode(out_seq, skip_special_tokens=True)
    # testing loop
    preds.extend(outs)
    tgts = tokenizer.batch_decode(batch_tgt, skip_special_tokens=True)
    targets.extend(tgts)
    progress_bar.update(1)


sacrebleu = evaluate.load("sacrebleu")


results = sacrebleu.compute(predictions=preds, references=targets)
print(results)

bleu1 = results['precisions'][0] * results['bp']
bleu2 = results['precisions'][1] * results['bp']
bleu3 = results['precisions'][2] * results['bp']
bleu4 = results['precisions'][3] * results['bp']

print("BLEU-1: ", bleu1)
print("BLEU-2: ", bleu2)
print("BLEU-3: ", bleu3)
print("BLEU-4: ", bleu4)
````

You can checkout all the available models here: https://huggingface.co/docs/transformers/en/model_doc/auto#transformers.AutoModelForSeq2SeqLM

````python
from transformers import AutoTokenizer

checkpoint = "google-t5/t5-small" #Select a model of your choice
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
````

We will need to tokenize both our input and outputs. Thus we make use of pre_process() function to generate tokenized model inputs and targets. Ensure you use truncation and padding! The max length will be 128.

````python
##Implement the preprocess function
def preprocess_function(examples):
    inputs = [example['fr'] for example in examples["translation"]]
    targets = [example['en'] for example in examples["translation"]]
    model_inputs = tokenizer(inputs, padding='max_length', truncation=True, max_length=128) #Instantitate tokenizer to generate model outputs
    labels = tokenizer(targets, padding='max_length', truncation=True, max_length=128) #Instantitate tokenizer to generate model outputs
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
````

````python
tokenized_train_data = train_dataset.map(preprocess_function, batched=True)
````

````python
tokenized_eval_data = eval_dataset.map(preprocess_function, batched=True)
````

We remove the column 'translation' as we do not require it for training. Also often having columns other than we created using the preprocess_function may lead to errors during training. Since model might get confused which inputs it needs to use.

````python
tokenized_train_data = tokenized_train_data.remove_columns(train_dataset.column_names)
tokenized_eval_data = tokenized_eval_data.remove_columns(eval_dataset.column_names)
````

````python
tokenized_train_data.set_format("torch")
tokenized_eval_data.set_format("torch")
````

To construct batches of training data for model training, we require collators that set the properties for the batches and data loaders that generate the batches.

````python
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model) #INSTANTIATE THE COLLATOR
````

````python
#Instantiate the DataLoader for training and evaluation data

from torch.utils.data import DataLoader

train_dataloader = DataLoader(tokenized_train_data, batch_size=8, shuffle=True, collate_fn=data_collator,) # Sending in collator here
eval_dataloader = DataLoader(tokenized_eval_data, batch_size=8, collate_fn=data_collator,)
````

## Q8) Training the Transformer Model

Now, that we have are data tokenized and ready in batches and model fixed. We will begin with training this model. To do so we must setup the right hyperparameters, then proceed to implment the training loop to train our model!

For training we require an optimizer and a scheduler to manage the learning rate during the training. Let's set them up before our training loop

````python
from torch.optim import AdamW
from transformers import get_scheduler

NUM_EPOCHS = 3
num_train_epochs = NUM_EPOCHS
num_training_steps = num_train_epochs * len(train_dataloader)

optimizer = AdamW(model.parameters(), lr=0.001)
# controls the learning rate
lr_scheduler = get_scheduler("linear", optimizer = optimizer, num_warmup_steps = 0, num_training_steps = num_training_steps)

````

````python
model.to('cuda')
````

Finally, we are here!

In the loop during training you will run a forward pass, compute the loss, compute the gradients, and then update the weights. (Don't foregt to set gradient to zero!)

During the eval phase we simply do a forward pass and compute the loss!

````python
from tqdm.auto import tqdm

progress_bar = tqdm(total=num_training_steps, desc="Training Progress")

for epoch in range(num_train_epochs):
    # Training Phase
    model.train()
    for batch in eval_dataloader:
        batch = {k: v.to('cuda') for k, v in batch.items()}

        model_output = model(**batch)
        # Model automatically calculates loss
        loss = model_output.loss
        # Backward Pass
        loss.backward()
        # Controlling the learning rate
        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)

    # Evaluation Phase
    model.eval()
    total_loss = 0
    total_batches = 0
    with torch.no_grad():
      for batch in eval_dataloader:
        batch = {k: v.to('cuda') for k, v in batch.items()}
        eval_outputs = model(**batch)
        logits = eval_outputs.logits
        labels = batch['labels']
        ### Complete the evaluation phase
        loss = eval_outputs.loss
        total_batches += 1
        total_loss += loss.item()

    avg_loss = total_loss/total_batches
    print(f"Epoch {epoch}: Average Eval Loss: {avg_loss:.4f}")
````
